```python
exac = pd.read_csv('../../../@broad/redux-170620-exac-daily-pageviews.csv', skiprows=6)
gnomad = pd.read_csv('../../../@broad/redux-170620-gnomad-daily-pageviews.csv', skiprows=6)

def format_google_csv(raw_df, name):
    dates = pd.to_datetime(raw_df['Day Index'])
    df = pd.DataFrame({ 
        'date'.format(name): dates, 
        '{}_pageviews'.format(name): raw_df['Pageviews'].map(lambda x: int(x.replace(',', ''))) 
    })
    # return df.set_index(df.date)student
    return df

exac_df = format_google_csv(exac, 'exac')
gnomad_df = format_google_csv(gnomad, 'gnomad')
combined = pd.merge_ordered(gnomad_df, exac_df, fill_method='zero')
combined = combined.fillna(value=int(0))
combined = combined.set_index(combined.date)
combined = combined['2014-10-13':'2017-06-09']
combined.to_json('../../../@broad/redux-170620-combine-pageviews.json', orient='records', date_format='iso')

```

```python
weeks = combined.resample('W').sum()
weeks.reset_index().to_json('../../../@broad/redux-170620-combine-pageviews-weeks.json', orient='records', date_format='iso')
```
